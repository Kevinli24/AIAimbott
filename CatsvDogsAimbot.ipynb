```python
import cv2
import numpy as np
import pyautogui
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array
```

```python
import os

file_path = "C:/Users/Chess/Downloads/dog_cat_classifier.h5"
print("File exists:", os.path.exists(file_path))
```

```python
# Load trained model
model = load_model("C:/Users/Chess/Downloads/dog_cat_classifier.h5")
class_names = ["Cat", "Dog"]
```

```python
def capture_screen():
    """Captures the screen and returns an OpenCV image."""
    screenshot = pyautogui.screenshot()
    frame = np.array(screenshot)
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
    return frame
```

```python
def detect_object(frame):
    """Detects whether a cat or dog is in the frame and returns its position."""
    img_resized = cv2.resize(frame, (150, 150))  # Resize to match model input
    img_array = img_to_array(img_resized) / 255.0
    img_array = np.expand_dims(img_array, axis=0)  # Expand dimensions

    predictions = model.predict(img_array)
    label_index = np.argmax(predictions)
    label = class_names[label_index]

    if predictions[0][label_index] > 0.7:  # Confidence threshold
        return label
    return None
```

```python
def find_and_click():
    """Captures screen, detects object, and clicks if found."""
    frame = capture_screen()
    label = detect_object(frame)

    if label:
        h, w, _ = frame.shape
        x, y = w // 2, h // 2  # Click center of the screen
        pyautogui.moveTo(x, y, duration=0.2)
        pyautogui.click()
        print(f"Clicked on {label}!")

# Run in a loop
while True:
    find_and_click()
```
